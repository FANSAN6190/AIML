```python
import cv2
cap = cv2.VideoCapture(0)
while(cap.isOpened()):
    ret,frame = cap.read()
    if ret == True:
        cv2.imshow('frame', frame)
    if cv2.waitKey(0) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
img = cv2.imread('elephant.jpg',100)
cv2.imshow('Original Image', img)
cv2.waitKey(0)
blurImg = cv2.blur(img,(10,10))
cv2.imshow('blurred image',blurImg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = np.random.randint(255, size=(300, 600, 3))

isWritten = cv2.imwrite('random_img.png', img)

if isWritten:
    print('The image is successfully saved.')
    img=cv2.imread("random_img.png",1)
    cv2.imshow("random_img",img)
    cv2.waitKey(0)
else:
    print("Failed")
```
