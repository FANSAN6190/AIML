```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#ploting a graph
a=[7,6,4,5,6,7,9,8]
plt.plot(a)
plt.show()
```

```python
y=[2,5,7,6]
x=[1,2,3,4]
plt.scatter(x,y)
plt.show()
```

```python
n=50
x=np.random.rand(n)
y=np.random.rand(n)
colors=np.random.rand(n)
area=np.pi * (15 * np.random.rand(n)**2)
plt.scatter(x,y,s=area,c=colors,alpha=0.5)
plt.show()
```

```python
#ploting an equation
m=10
b=5
x=np.array([1,8,23,34])
y=m*x+b
plt.plot(x,y)
plt.show()
```

```python
#styling graph
#adding colors and markers
x=np.array([1,2,3,4,5])
y=np.array([23,27,31,40,37])
plt.plot(x,y,color="orange",marker="o")
plt.show()
```

```python
#adding linewidth and special markers
x=np.array([1,2,3,4,5])
y=np.array([23,27,31,40,37])
plt.plot(x,y,"b--o",linewidth=0.5) #marker=k^:,bo,b+,b--,g--d
plt.show()
```

```python
#adding labels and title
x=np.array([1,2,3,4,5])
y=np.array([23,27,31,40,37])
plt.plot(x,y)
plt.ylabel("Random Values")
plt.xlabel("Ponters")
plt.title("Ploting a Random Graphs")
plt.show()
```

```python
#modifying axis value range
plt.axis([0,6,0,50])
x=np.array([1,2,3,4,5])
y=np.array([23,27,31,40,37])
plt.plot(x,y)
plt.show()
```

```python
#graph with grid
plt.grid(True)
plt.axis([0,6,0,50])
x=np.array([1,2,3,4,5])
y=np.array([23,27,31,40,37])
plt.plot(x,y)
plt.show()
```

```python
#axis scaling (linear,logarithmic)
x=np.array([1,2,3,4,5])
y=np.array([23,27,31,40,37])
plt.xscale('linear')
plt.yscale('log')
plt.plot(x,y)
plt.show()
```
